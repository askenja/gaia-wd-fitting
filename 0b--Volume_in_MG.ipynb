## Modeled volume

I model the following volume: 

- |R - RSun| < 150 pc (Rsun = 8.275 kpc)
- |z| < 150 pc
- d_max = 1 kpc

As I select data with 2 mag < G < 20.5 mag, different stars have different min-max distances
at which they can be observed within this G range (see notebook 0a). So d_max can be less than 
1 kpc for some stars. 

Here I calculate volume as a function of z of such a selection. 
Volume depends on the absolute magnitude M_G. 

```python
import numpy as np
import matplotlib.pyplot as plt
from jjmodel.geometry import Volume

from jjmodel.input_ import p, a
```

```python
# Load z grid from the input module
z_grid = a.z  # pc

# Define distance limits for this study
z_max = 150   # pc 
dR = 0.15     # kpc
R_Sun = 8.275 # kpc (also in parameter file)
d_max = 1000  # pc  

# And create distance grid
d_bins = np.arange(1, d_max + 1, 1)  # pc
```

```python
# Calculate volumes for z and d_max from the distance grid 
# My volume function has been already defined for SJ21 paper and describes 
# excactly this type of volume - sphere cut by Galactocentric annuli

v = Volume(p,a)

vz_grid = np.zeros((len(z_grid),len(d_bins)))
for i, d in enumerate(d_bins):
    vz_grid[:,i] = v._circ_volume_(0, d, R_Sun - dR, R_Sun + dR)
```

```python
# Index 30 corresponds to z_max = 150 pc with z_step = 5 pc 
# plt.plot(z_grid[:30],vz_grid[:30,88])

np.savetxt('./completeness/vz_grid_myvolume.txt', vz_grid[:30,:],
          header='Volume: local sphere of radius d_max + cut RSun - dR < R < RSun + dR, dR = 150 pc' +\
          '\nRows: bin centers of z_grid = (1,150) pc, step 5 pc; Columns: d_max = (1, 1000) pc, step 1 pc')
```
