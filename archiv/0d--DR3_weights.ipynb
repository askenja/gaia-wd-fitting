```python
import numpy as np 
```

```python
r_sphere = 600  # pc
dR = 150        # pc
Rsun = 8.17*1e3 # pc

R1 = Rsun - dR
R2 = Rsun + dR
```

```python
def d_minmax(l,b,R1,R2,R0=8170):

    l_rad = np.deg2rad(l)
    b_rad = np.deg2rad(b)

    d11 = - R0*np.cos(l_rad)/np.cos(b_rad) + np.sqrt(R1**2 - np.sin(l_rad)**2*R0**2)/np.cos(b_rad)
    d12 = - R0*np.cos(l_rad)/np.cos(b_rad) - np.sqrt(R1**2 - np.sin(l_rad)**2*R0**2)/np.cos(b_rad)

    d21 = - R0*np.cos(l_rad)/np.cos(b_rad) + np.sqrt(R2**2 - np.sin(l_rad)**2*R0**2)/np.cos(b_rad)
    d22 = - R0*np.cos(l_rad)/np.cos(b_rad) - np.sqrt(R2**2 - np.sin(l_rad)**2*R0**2)/np.cos(b_rad)

    return d11, d12, d21, d22


def roots_for_Rtarget(R_targ, l_rad, b_rad, R0):
    A = (np.cos(b_rad))**2
    B = -2.0 * R0 * np.cos(b_rad) * np.cos(l_rad)
    C = R0**2 - R_targ**2
    disc = B*B - 4*A*C
    if disc < 0:
        return []
    r1 = (-B - np.sqrt(disc)) / (2*A)
    r2 = (-B + np.sqrt(disc)) / (2*A)
    roots = sorted([r1, r2])
    return roots


def choose_roots(roots_r1,roots_r2,roots_sph):

    roots = [roots_sph]

    # choose min positive value
    roots_r1_pos = [r for r in roots_r1 if r > 0]
    roots_r2_pos = [r for r in roots_r2 if r > 0] 

    print(roots_r1_pos, roots_r2_pos)

    if roots_r1_pos != []:
        roots += [min(roots_r1_pos)]
    
    if roots_r2_pos != []:
        roots += [min(roots_r2_pos)]

    return min(roots)

```

```python
l = 90
b = 0

l_rad = np.deg2rad(l)
b_rad = np.deg2rad(b)

#d = d_minmax(l,b,R1,R2,R0=Rsun)
#print(d)

r1 = roots_for_Rtarget(R1, l_rad, b_rad, Rsun)
r2 = roots_for_Rtarget(R2, l_rad, b_rad, Rsun)
print(r1, r2)

r = choose_roots(r1,r2,r_sphere)
print(1000/r)
```
