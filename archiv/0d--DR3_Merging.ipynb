```python
import pyarrow.dataset as ds
import pyarrow.parquet as pq
```

```python
def combine_parquet_folder(input_dir, output_file):
    dataset = ds.dataset(input_dir, format="parquet")
    table = dataset.to_table()       # efficient + parallel + low memory
    pq.write_table(table, output_file)

combine_parquet_folder("./data/600pc_clean", "./data/600pc_clean/combined.parquet")
```
