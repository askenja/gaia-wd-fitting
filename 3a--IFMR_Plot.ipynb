```python
import numpy as np 
import matplotlib.pyplot as plt

from helpers import IFMRHandler
```

```python
ifmr = IFMRHandler()

x = np.linspace(0.87, 8.2, 500)
y, sigma_y = ifmr.cummings(x)

plt.figure(figsize=(8,6))
plt.plot(x, y, color="black", lw=2)
# Plot shaded uncertainty bands
for n, alpha in zip([1,2,3], [0.3, 0.2, 0.1]):
    plt.fill_between(x, y - n*sigma_y, y + n*sigma_y, color="blue", alpha=alpha)

# Labels and styling
plt.xlabel(r"Initial mass $M_{\mathrm{ini}}$ [$M_{\odot}$]", fontsize=14)
plt.ylabel(r"Final mass $M_f$ [$M_{\odot}$]", fontsize=14)
plt.title("WD IFMR Cummings+18 (forward)", fontsize=15)
plt.grid(True, ls="--", alpha=0.6)
plt.tight_layout()
plt.savefig('IFMR_cummings_parsec.png')


x = np.linspace(0.75, 1.25, 500)
y, sigma_y = ifmr.cummings_r(x)
plt.figure(figsize=(8,6))
plt.plot(x, y, color="black", lw=2)
# Plot shaded uncertainty bands
for n, alpha in zip([1,2,3], [0.3, 0.2, 0.1]):
    plt.fill_between(x, y - n*sigma_y, y + n*sigma_y, color="blue", alpha=alpha)

# Labels and styling
plt.ylabel(r"Initial mass $M_{\mathrm{ini}}$ [$M_{\odot}$]", fontsize=14)
plt.xlabel(r"Final mass $M_f$ [$M_{\odot}$]", fontsize=14)
plt.title("WD IFMR Cummings+18 (inverted)", fontsize=15)
plt.grid(True, ls="--", alpha=0.6)
plt.tight_layout()
plt.savefig('IFMR_inverted_cummings_parsec.png')
```
